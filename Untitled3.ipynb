```python
X = train.drop(['price_range'], axis=1)
```

```python
y = train['price_range']
```

```python
X['touch_screen'] = np.where(X['touch_screen'] == '-', 'Yes', X['touch_screen'])
```

```python
test['touch_screen'] = np.where(test['touch_screen'] == '-', 'Yes', test['touch_screen'])
```

```python
X['touch_screen'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
categorical_columns = ['touch_screen']
X_encoded = pd.get_dummies(X, columns=categorical_columns, drop_first=True)

encoder = OneHotEncoder(sparse_output=False)

one_hot_encoded = encoder.fit_transform(X[categorical_columns])

one_hot_df = pd.DataFrame(one_hot_encoded,
                          columns=encoder.get_feature_names_out(categorical_columns))

df_sklearn_encoded = pd.concat([X.drop(categorical_columns, axis=1), one_hot_df], axis=1)

df_sklearn_encoded
```

```python
one_hot_encoded = encoder.fit_transform(test[categorical_columns])

one_hot_df = pd.DataFrame(one_hot_encoded,
                          columns=encoder.get_feature_names_out(categorical_columns))

test_enc = pd.concat([test.drop(categorical_columns, axis=1), one_hot_df], axis=1)

test_enc
```
